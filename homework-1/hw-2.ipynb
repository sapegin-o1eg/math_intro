### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x = []
y1 = []
y2 = []

p = 5
for i in range(1000):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt(2*p*x_))
    y2.append(-np.sqrt(2*p*x_))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

## Парабола

```python
x = np.linspace(0,50, 1000)
y1 = np.sqrt(2*p*x)
y2 = -np.sqrt(2*p*x)
plt.axhline(linewidth=0.2, color='black')
plt.axvline(linewidth=0.2, color='black')
plt.axis('equal')
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

## Окружность 

```python
r=1
x = np.linspace(-1,1, 2000)
y = np.sqrt(r**2-x**2)
plt.axhline(linewidth=0.2, color='black')
plt.axvline(linewidth=0.2, color='black')
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, -y)
plt.xlabel('x')
plt.ylabel('y')
```

## Эллипс 

```python
a = 3
b = 2
x = np.linspace(-a,a, 2000)
y = np.sqrt(b**2*(1-x**2/a**2))
plt.axhline(linewidth=0.2, color='black')
plt.axvline(linewidth=0.2, color='black')
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, -y)
plt.xlabel('x')
plt.ylabel('y')
```

## Гипербола 

```python
a = 1
b = 2
x = np.linspace(a, a+3, 1000)
y = np.sqrt(b**2 * (x**2/a**2 - 1))
plt.axhline(linewidth=0.2, color='black')
plt.axvline(linewidth=0.2, color='black')
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, -y)
plt.plot(-x, y)
plt.plot(-x, -y)
plt.xlabel('x')
plt.ylabel('y')
```
