```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def cos_kx(k,x):
    """
    returns f(k,x) = cos(kx)
    """
    return np.cos(k*x)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
plt.figure(figsize=(10,5))
plt.plot(x, cos_kx(1,x), label='cos(x)')
plt.plot(x, cos_kx(2,x), label='cos(2*x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```
